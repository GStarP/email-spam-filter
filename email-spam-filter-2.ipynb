```python
import os
import numpy as np

# 数据目录(如要运行代码请自行修改!!!)
DATA_DIR = 'D:\MY\ml-data\Enron-Spam\pre-processed'
# 划分训练集(可自由组合 1~6 六个目录)(测试集见最后一个代码单元)
TRAIN = [1, 2, 3, 4, 5]
TRAIN_DIRS = []
for i in TRAIN:
    TRAIN_DIRS.append(os.path.join(DATA_DIR, 'enron' + str(i)))

# 提取样本信息
def build_sample(data_dirs):
    texts = []
    labels = []
    for d in data_dirs:
        ham_d = os.path.join(d, 'ham')
        for f in os.listdir(ham_d):
            fd = os.path.join(ham_d, f)
            # 0-ham, 1-spam
            labels.append(0)
            with open(fd, encoding='utf-8', errors='ignore') as email:
                next(email)
                text = email.read().replace('\n', ' ')
                texts.append(text)
    for d in data_dirs:
        spam_d = os.path.join(d, 'spam')
        for f in os.listdir(spam_d):
            fd = os.path.join(spam_d, f)
            labels.append(1)
            with open(fd, encoding='utf-8', errors='ignore') as email:
                next(email)
                text = email.read().replace('\n', ' ')
                texts.append(text)
    # 返回邮件样本列表和邮件标签列表
    return texts, labels

train_text, train_labels = build_sample(TRAIN_DIRS)
print(train_labels)
print(train_text[1])
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# 进行词频统计
print('===== Counter =====')
cv = CountVectorizer(stop_words='english', max_df=0.3, decode_error='ignore')
train_counts = cv.fit_transform(train_text)
print(train_counts)
print(train_counts.shape)
# 通过 TF-IDF 对邮件内容进行特征提取
print('===== TF-IDF =====')
tf = TfidfTransformer()
train_matrix = tf.fit(train_counts).transform(train_counts)
print(train_matrix)
print(train_matrix.shape)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, roc_auc_score
import pandas as pd

# 训练模型
model = LinearSVC()
model.fit(train_matrix, train_labels)

# 用同样的方式处理测试集
TEST = [6]
TEST_DIRS = []
for i in TEST:
    TEST_DIRS.append(os.path.join(DATA_DIR, 'enron' + str(i)))
test_text, test_labels = build_sample(TEST_DIRS)
cv2 = CountVectorizer(stop_words='english', max_df=0.3, decode_error='ignore', vocabulary=cv.vocabulary_)
test_counts = cv2.fit_transform(test_text)
# 由于使用了 cv 的词汇表, 测试集的词频向量与训练集是同维的
print(test_counts.shape)
test_matrix = tf.fit(test_counts).transform(test_counts)

# 预测结果并评估
res = model.predict(test_matrix)
rm = pd.DataFrame(
    confusion_matrix(test_labels, res),
    index=['ham','spam'],
    columns=['ham','spam']
)
print(rm)
print('precision score: %f' % precision_score(test_labels, res))
print('recall score: %f' % recall_score(test_labels, res))
# 引入了另外两种评估手段: 精度, ROC 曲线下面积
print('accuracy score: %f' % accuracy_score(test_labels, res))
print('roc-auc score: %f' % roc_auc_score(test_labels, res))
```
